```python
import requests

# Reemplaza con tu clave API
API_KEY = "5ae2e3f221c38a28845f05b609e9af6742b0197e80e4d22bad91036f"

def obtener_sitios_turisticos_ciudad():
    """
    Obtiene una lista de sitios turísticos de interés en Madrid, filtrados por popularidad.
    """
    # Define la URL del endpoint de la API para obtener lugares en una ciudad
    url = "https://api.opentripmap.com/0.1/en/places/bbox"
    
    # Configura los parámetros de la búsqueda
    params = {
        "lat_min": 40.3,  # Latitud mínima de la ciudad
        "lat_max": 40.7,  # Latitud máxima de la ciudad
        "lon_min": -3.8,  # Longitud mínima de la ciudad
        "lon_max": -3.4,  # Longitud máxima de la ciudad
        "apikey": API_KEY,  # Tu clave API
        "limit": 40,  # Limita el número de lugares de interés a 20
        "rate": 1  # Solo lugares muy relevantes y famosos (1 = más relevantes, 3 = menos relevantes)
    }
    
    # Realiza la solicitud GET
    response = requests.get(url, params=params)
    
    # Verifica si la solicitud fue exitosa (código 200)
    if response.status_code == 200:
        lugares = response.json().get("features", [])
        # Retorna los nombres de los sitios encontrados
        return [lugar["properties"]["name"] for lugar in lugares if lugar["properties"].get("name")]
    else:
        print(f"Error al obtener los sitios de interés. Código de estado: {response.status_code}")
        return []

def main():
    # Llamada a la función que obtiene los sitios turísticos en Madrid
    sitios = obtener_sitios_turisticos_ciudad()
    
    if sitios:
        print(f"Sitios turísticos en Madrid:")
        for sitio in sitios:
            print(f"- {sitio}")
    else:
        print("No se encontraron sitios turísticos.")

if __name__ == "__main__":
    main()
```
